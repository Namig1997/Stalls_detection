```python
import plotly.express as px
from plotly.offline import plot
import pandas as pd
import numpy as np
```

```python
logs = pd.read_csv("../logs/logs_cnn_8_filters_4_layers_lr_slight_decay_max_pooling.csv")
logs['iteration'] = logs.index.values
logs_melt = logs.melt(id_vars='iteration', value_vars=['lr', 'loss'])
fig = px.line(logs_melt, x='iteration', y='value', color='variable')
fig.show()
```

```python
metrics = pd.read_csv("../logs/metrics_logs_cnn_8_filters_4_layers_lr_slight_decay.csv")
metrics['epoch'] = metrics.index.values
metrics_melt = metrics.melt(id_vars='epoch', value_vars=['mcc', 'accuracy', 'precision', 'recall', 'f1', 'auc'])
fig = px.line(metrics_melt, x='epoch' , y='value' , color='variable')
fig.show()
```
